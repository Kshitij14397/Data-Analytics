```
import matplotlib.pyplot as plt
import numpy as np
```

```
print('Enter the x and y value of a vector')
x1 = float(input())
y1 = float(input())
print('Original vector: ({0}, {1})'.format(x1, y1))
```

```
print('Enter the scalar quantity')
c = float(input())
x2 = c*x1
y2 = c*y1
print('Resultant vector: ({0}, {1})'.format(x2, y2))
```

```
ax = plt.axes()
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D vectors plot')
plt.grid()
ax.arrow(0, 0, x1, y1, head_width=0.5, head_length=0.7, fc='green', ec='black')
ax.arrow(0, 0, x2, y2, head_width=0.5, head_length=0.7, fc='blue', ec='black')
plt.show()
```
