```
import matplotlib.pyplot as plt
import numpy as np
```

```
print('Enter the x and y value of 1st vector')
x1 = float(input())
y1 = float(input())
print('1st vector i.e a: ({0}, {1})'.format(x1, y1))
```

```
print('Enter the x and y values of 2nd vector')
x2 = float(input())
y2 = float(input())
print('2nd vector i.e b: ({0}, {1})'.format(x2, y2))
```

```
print('Enter the starting value(scalar quantity) of range')
p = int(input())
print('Enter the ending value(scalar quantity) of range')
q = int(input())
```

```
ax = plt.axes()
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D vectors plot')
plt.grid()

ax.arrow(0, 0, x1, y1, head_width=0.5, head_length=0.7, fc='red', ec='black')
ax.arrow(0, 0, x2, y2, head_width=0.5, head_length=0.7, fc='red', ec='black')

for i in range(p, (q+1)):
  x = x1 + (i * x2)
  y = y1 + (i * y2)
  ax.arrow(0, 0, x, y, head_width=0.5, head_length=0.7, fc='red', ec='black')
plt.show()

```
