```
import matplotlib.pyplot as plt
import numpy as np
```

```
print('Enter the x and y value of 1st vector')
x1 = float(input())
y1 = float(input())
print('1st vector: ({0}, {1})'.format(x1, y1))
```

```
print('Enter the x and y values of 2nd vector')
x2 = float(input())
y2 = float(input())
print('2nd vector: ({0}, {1})'.format(x2, y2))
```

```
print('After addition,')
x3 = x1 + x2
y3 = y1 + y2
print('3rd vector: ({0}, {1})'.format(x3, y3))
```

```
ax = plt.axes()
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D vectors plot')
plt.grid()
ax.arrow(0, 0, x1, y1, head_width=0.5, head_length=0.7, fc='green', ec='black')
ax.arrow(0, 0, x2, y2, head_width=0.5, head_length=0.7, fc='blue', ec='black')
ax.arrow(0, 0, x3, y3, head_width=0.5, head_length=0.7, fc='red', ec='black')
plt.show()
```
